### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
# Agrupando os nascimentos por nome do municipio e verificando qual é o municipio com menos nascimentos
sinasc_raw.groupby('munResNome')['DTNASC'].count().sort_values().head(5)
```

```python
# Convertendo DTNASC para datetime e extraindo o ano
sinasc_raw['DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])
sinasc_raw['ANO_NASC'] = sinasc_raw['DTNASC'].dt.year

# Filtrando para o ano de 2019
sinasc_2019 = sinasc_raw[sinasc_raw['ANO_NASC'] == 2019]

# Agrupando por município e contando os nascimentos
nascimentos_por_municipio = sinasc_2019.groupby('munResNome')['DTNASC'].count().sort_values()

# Encontrando o município com menos nascimentos
municipio_menos_nascimentos = nascimentos_por_municipio.index[0]

# Filtrando e agregando dados para o município com menos nascimentos
sinasc_menor_municipio = sinasc_2019[sinasc_2019['munResNome'] == municipio_menos_nascimentos]
estatisticas_idades = sinasc_menor_municipio.groupby('munResNome')[['IDADEMAE', 'IDADEPAI']].agg(['mean', 'max', 'min'])

print(f"O município com menos nascimentos em 2019 é: {municipio_menos_nascimentos}")
print(estatisticas_idades)
```

```python
def analisar_municipio(df, municipio):
    """
    Analisa os dados de nascimentos para um município específico.

    Args:
        df (pd.DataFrame): DataFrame com os dados de nascimentos.
        municipio (str): Nome do município a ser analisado.

    Returns:
        pd.DataFrame: DataFrame com as estatísticas calculadas.
    """

    # Filtra os registros do município informado
    municipio_df = df[df['munResNome'] == municipio]

    # Verifica se o município existe nos dados
    if municipio_df.empty:
        print(f"Município '{municipio}' não encontrado nos dados.")
        return None

    # Agrupa os dados e calcula as estatísticas
    resultados = municipio_df.groupby('munResNome').agg({
        'QTDFILVIVO': ['min', 'mean', 'max'],
        'QTDFILMORT': ['min', 'mean', 'max'],
        'IDADEMAE': ['min', 'mean', 'max'],
        'IDADEPAI': ['min', 'mean', 'max']
    })

    return resultados
```
